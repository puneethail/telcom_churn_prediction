```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('cleaned.csv')
```

```python
df.churn.value_counts()
```

```python
sns.histplot(df['tenure'], kde=True)  # kde=True overlays a density curve
plt.title('Monthly Charges Distribution')
plt.show()
```

```python
# Plot countplot to visualize Partner vs Churn
plt.figure(figsize=(7, 4))
sns.countplot(data=df, x='partner', hue='churn', palette='Set2')
plt.title('Partner vs Churn')
plt.xlabel('Has Partner?')
plt.ylabel('Number of Customers')
plt.legend(title='Churn')
plt.tight_layout()
plt.show()
```

```python
# Plot countplot to visualize Partner vs Churn
plt.figure(figsize=(7, 4))
sns.countplot(data=df, x='dependents', hue='churn', palette='Set2')
plt.title('Dependent vs Churn')
plt.xlabel('Dependent?')
plt.ylabel('Number of Customers')
plt.legend(title='Churn')
plt.tight_layout()
plt.show()
```

```python
def group_tenure(value):
    if value>=0 and value<=12:
        return 0
    elif value>=13 and value<=24:
        return 1
    elif value>=25 and value<=36:
        return 2
    elif value>=37 and value<=48:
        return 3
    elif value>=49 and value<=60:
        return 4
    elif value>=61 and value<=72:
        return 5
    elif value>72:
        return 6
```

```python
df['tenure_group'] = df['tenure'].map(group_tenure)
```
